```python
import tensorflow as tf
from tensorflow import keras # FOR MAKING NEURAL NETWORKS
import matplotlib.pyplot as plt
```

**Loading the data set**
MNIST: Handwritten digits dataset (0-9) with 28x28 grayscale images, used for training neural networks.

```python
(X_train, y_train),(X_test, y_test)= keras.datasets.mnist.load_data()
```

**FLATTENING THE DATA SET**
28 X 28 Image -> Single dimensional Array

```python
X_train.shape
```

```python
X_train= X_train/255
X_test= X_test/255
```

```python
X_train_flattened = X_train.reshape(len(X_train),28*28)
```

```python
X_train_flattened.shape
```

```python
X_test_flattened= X_test.reshape(len(X_test),28*28)
```

```python
X_test_flattened.shape
```

**CREATING SIMPLE NUERAL NETWORK**
1 input and 1 output layer

```python
from tensorflow.keras import layers
```

```python
model = keras.Sequential([
    layers.Dense(10, input_shape=(784,), activation='sigmoid') #10 output neurons, 784 input features
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train_flattened, y_train,epochs=5)
```

```python
model.evaluate(X_test_flattened, y_test)
```

**SAMPLE PREDICTION**

```python
plt.matshow(X_test[0])
```

```python
y_predicted=model.predict(X_test_flattened)
y_predicted[0]
```
